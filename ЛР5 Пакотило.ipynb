### 2.9. создание нового поля вычисленного на основе значений других полей:

#### 2.9.1 через выражение на базе имеющихся колонок,

```python
new = [rests.Name.sample(n=1, random_state=1), rests.Location.sample(n=1, random_state=1), rests.Locality.sample(n=1, random_state=1), rests.City.sample(n=1, random_state=1), rests.Cuisine.sample(n=1, random_state=1), rests.Rating.mean().round(), rests.Votes.mean().round(), rests.Cost.mean().round()]
rests.append(pd.Series (new, index=rests.columns [: len (new)]), ignore_index= True )
#rests.Name.sample(n=1, random_state=1)
```

```python
rests['CITIES'] = rests['City'].apply(lambda element: element.upper())

rests
```
